La manera de manejar los datos de dicha fecha puede varias según la importancia de los mismos. Si tenemos sufucientes datos para lo que queremos y los erróneos no son clave, simplemente podemos eliminar las filas incorrectas. En caso de que no queramos/podamos elimina dichas filas, lo mejor que podemos hacer es imputar los datos. Si no tenemos referencia alguna sobre lo que ha fallado, simplemente podemos rellenar con la media de cada columna. Si estamos seguros de que el ticker está correcto, podemos rellenar utilizando la media de cada ticker. Además si solo falla una de las columnas, siempre podemos usar las otras que si están correctas para intentar inferir un valor aproximado a partir de otras filas con datos similares.
