### Paso 3. Asígnalo a una variable llamada usuarios.

```python
usuarios = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep="|")
usuarios.head()
```

### Paso 4. Descubre cuál es la edad media por ocupación

```python
usuarios.groupby(["occupation"], as_index=False)["age"].mean()
```

### Paso 5. Descubra la proporción de Hombres por ocupación y ordénela de mayor a menor

#### Forma 1

```python
F_1 = usuarios.groupby("occupation")["gender"].value_counts().unstack().fillna(0)
F_1
```

```python
F_1["P_M"] = F_1["M"]/(F_1["M"]+F_1["F"])
F_1
```

```python
F_1["P_M"].sort_values(ascending=False).reset_index()
```

#### Forma 2

```python
F_2 = usuarios[usuarios.columns[2:4]].value_counts().unstack().T.fillna(0)
F_2
```

```python
F_2["P_M"] = F_2["M"]/(F_2["M"]+F_2["F"])
```

```python
F_2["P_M"].sort_values(ascending=False).reset_index()
```

### Paso 6. Para cada ocupación, calcule las edades mínima y máxima

```python
usuarios.groupby("occupation")["age"].agg(["max", "min"])
```

### Paso 7. Para cada combinación de ocupación y sexo, calcule la edad media

```python
usuarios.groupby(["gender","occupation"])["age"].mean().unstack().T
```

### Paso 8.  Para cada ocupación, presente el porcentaje de mujeres y hombres

```python
# Para hombres utilizamos la proporcion utilizada anteriormente
F_1["P_M"]*100
```

```python
# Para mujeres 
P_F = usuarios.groupby("occupation")["gender"].value_counts().unstack().fillna(0)
```

```python
pd.concat([(P_F["F"]/(P_F["M"]+P_F["F"]))*100, F_1["P_M"]*100], axis=1).rename(columns= {0:"Porcentaje mujeres", "P_M":"Porcentaje hombres"})
```
